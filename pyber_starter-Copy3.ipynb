```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_data_read=pd.read_csv(city_data_to_load)
ride_data_read=pd.read_csv(ride_data_to_load)
#city_data_read.head()

# Combine the data into a single dataset
city_ride_merge=pd.merge(ride_data_read,city_data_read, on="city", how = "left")

# Display the data table for preview
city_ride_merge.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
rides_per_city=city_ride_merge.groupby('city')['ride_id'].count()
xaxis_limit=rides_per_city.max()
#xaxis
avg_fare=city_ride_merge.groupby('city')['fare'].mean()
yaxis_limit=avg_fare.max()
#yaxis
driverspercity=city_ride_merge.groupby('city')['driver_count'].sum()
driverspercity

# Build the scatter plots for each city types
labels='Rural', 'Suburban','Urban'
sizes = driverspercity


#colors={'Rural':'gold','Suburban':'lightskyblue','Urban':'lightcoral'}

plt.scatter(rides_per_city, avg_fare, marker="o", facecolors=('lightskyblue'), edgecolors="black",
            s=((driverspercity)/3), alpha=0.8)

# Incorporate the other graph properties

plt.title('Pyber Ride Sharing Data (2016)')
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Avg. Fares ($)")
plt.grid(True)

# Create a legend
plt.legend()


# Incorporate a text label regarding circle size

# Save Figure
plt.savefig('Pyber_Ride_Sharing_Data(2016).png') 
```

```python
# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents

faretotalpercity=city_ride_merge.groupby('type')['fare'].sum()
#faretotalpercity
totalfare=faretotalpercity.sum()
#totalfare
#faretotalpercity


#
# Build Pie Chart
labels='Rural', 'Suburban','Urban'
sizes = faretotalpercity
colors = ['gold','lightskyblue','lightcoral']
explode= [0,0,0.10]
#Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=150)
#plt.legend(labels, loc="upper right")
#plt.tight_layout()
plt.title('% of Total Fares by City Type')
plt.axis('auto')

plt.show()
# Save Figure
plt.savefig('%_of_Total_Fares_by_City_Type.png') 
```

```python
# Show Figure
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents
totalridespercitytype=city_ride_merge.groupby('type')['ride_id'].count()

# Build Pie Chart
labels='Rural', 'Suburban','Urban'
sizes = totalridespercitytype
colors = ['gold','lightskyblue','lightcoral']
explode= [0,0,0.10]
#Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=150)
plt.title('% of Total Rides by City Type') 
plt.axis('auto')
plt.show() 
# Save Figure
plt.savefig('%_of_Total_Rides_by_City_Type.png') 
```

```python
# Show Figure
plt.show()
```
